```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt
```

```python
n = 20
input_sizes = np.arange(1,n+1)
input_sizes = input_sizes.astype(int)

k_time = np.ones(n) # O(1)
log_time = np.log2(input_sizes) #O(log n)
linear_time = input_sizes #O(n)

linearithmic_time = input_sizes * np.log2(input_sizes) #O(n log n)
quadratic_time = input_sizes**2 #O(n^2)
polynomial_time = input_sizes**3 #O(n^k), where k is a constant; in this case O(n^3)

exponential_time = 2**input_sizes #O(2^n)
k_exponential_time = 3**input_sizes #O(k^n), where k is a constant; in this case O(3^n)
factorial_time = scipy.special.factorial(input_sizes) #O(n!)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.plot(input_sizes,k_time,'black')
plt.plot(input_sizes,log_time,'green')
plt.plot(input_sizes,linear_time,'blue')
plt.plot(input_sizes,linearithmic_time,'orange')
plt.plot(input_sizes,quadratic_time,'red')
plt.title('Visualising Runtime Complexities')
plt.ylabel('CPU Operations')
plt.xlabel('Input Sizes')

plt.subplot(1,2,2)
plt.yscale('log')
plt.plot(input_sizes,quadratic_time,'black')
plt.plot(input_sizes,polynomial_time,'green')
plt.plot(input_sizes,exponential_time,'blue')
plt.plot(input_sizes,k_exponential_time,'orange')
plt.plot(input_sizes,factorial_time,'red')
plt.title('Visualising Runtime Complexities')
plt.ylabel('CPU Operations (Log Scale)')
plt.xlabel('Input Sizes')

plt.show()
```

```python
## O(1)  Constant time
nums = [1,2,3,4,5,4,5,6,7,8]
print(nums) 
```

```python
## O(n)  Linear time
nums = [1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8]
for i in nums:
    print(i)
```

```python
## O(n^2)  Quadratic time Big-O squared
nums = [1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8]
for i in nums:
    for j in nums:
        print(i+j)
```

```python
nums = [1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8]
for i in nums:
    for j in nums:
        for k in nums[2:]:
            print(i+j)
```

```python
nums = [1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8,1,2,3,4,5,4,5,6,7,8]

for i in len(nums):
    count = 0
    while len(nums) > 1:
        count += 1
        nums = nums[:len(nums)//2]

print(count)
```
