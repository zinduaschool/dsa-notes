```python
## Factorial Function
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
def sum_(n):
    if n == 1:
        return 1
    return n + sum_(n-1)

sum_(100)
```

```python
cache = dict() ### Key= n, value (f(n))
def fibo(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    if n in cache.keys():
        return cache[n]
    cache[n] = fibo(n-1) + fibo(n-2)
    return cache[n]
```

```python
fibo(1000)
```

```python
def fibonacci(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    a = 1
    b = 1
    for i in range(2,n):
        c = a + b
        a = b
        b = c

    return b

fibonacci(10000)
```
